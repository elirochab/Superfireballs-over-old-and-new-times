```python
def datetoSolarL(date): # fechas --> 'long solar' medida en la ecliptica
    s =ep.Sun()         
    s.compute(date)     #Ejecutamos la función s para calcular las efemerides en una fecha concreta
    ecl = ep.Ecliptic(s) #Transformamos las coordenadas en coordenadas eclipticas (ascensión_recta-declinación)
    ecl.lon              #Nos da la longitud ecliptica
    #print ecl.lon
    solarL=(ecl.lon/np.pi)*180 #Pasamos de radianes a grados.
    return solarL
```

```python
def SolarLtoDate(y):
    year = int(y/360) # 1
    frac_year = y/360 - year # 2
    day_of_year = int(frac_year * 365.24) # 3
    table = {
        0: (1, 1), # mes: (día inicial, día final)
        31: (2, 28),
        59: (3, 31),
        90: (4, 30),
        120: (5, 31),
        151: (6, 30),
        181: (7, 31),
        212: (8, 31),
        243: (9, 30),
        273: (10, 31),
        304: (11, 30),
        334: (12, 31)
    }
    for days, (month, last_day) in table.items():
        if day_of_year <= days:
            break
    day = day_of_year - (days - last_day) # 4
    return "{}/{}/{}".format(day, month, year)
```
