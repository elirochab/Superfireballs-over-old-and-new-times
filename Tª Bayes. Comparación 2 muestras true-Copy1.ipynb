```python
def changes(string_list):
    reem=[]
    for string in string_list :
        reem_string=string.replace('-','/')
        reem.append(reem_string)
    return reem
    
```

```python
import ephem as ep #Ephemerides
import numpy as np 
import matplotlib.mlab as mlab 
import matplotlib.pyplot as plt 
import argparse
from datetime import datetime

def datetoSolarL(date): # fechas --> 'long solar' medida en la ecliptica
    s =ep.Sun()         
    s.compute(date)     #Ejecutamos la función s para calcular las efemerides en una fecha concreta
    ecl = ep.Ecliptic(s) #Transformamos las coordenadas en coordenadas eclipticas (ascensión_recta-declinación)
    ecl.lon              #Nos da la longitud ecliptica
    #print ecl.lon
    solarL=(ecl.lon/np.pi)*180 #Pasamos de radianes a grados.
    return solarL
```

```python
import ephem as ep
from datetime import date
import numpy as np
def datetoSolarL(date): # fechas --> 'long solar' medida en la ecliptica
    s =ep.Sun()         
    s.compute(date)     #Ejecutamos la función s para calcular las efemerides en una fecha concreta
    ecl = ep.Ecliptic(s) #Transformamos las coordenadas en coordenadas eclipticas (ascensión_recta-declinación)
    ecl.lon              #Nos da la longitud ecliptica
    #print ecl.lon
    solarL=(ecl.lon/np.pi)*180 #Pasamos de radianes a grados.
    return solarL
```

```python
import numpy as np
import pandas as pd
from numpy import std
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
#
proportion_values = np.arange(-46, 406, 0.433) #Del primer al último dato en pasos de 0.433, la primera muestra es la más 
#grande con 1044 datos, por lo que se divide el rango del intervalo entre estos datos y resulta 0.433
prior_kde = gaussian_kde(repllistnr, bw_method=0.025)
# compute the likelihood distribution as a KDE
likelihood_kde = gaussian_kde(replcneo, bw_method=0.025)
# compute the posterior distribution by multiplying the prior and likelihood and normalizing
posterior_densities = likelihood_kde(proportion_values) * prior_kde(proportion_values)
posterior_densities /= posterior_densities.sum()

# plot the prior, likelihood, and posterior distributions
fig, ax = plt.subplots()
fig.set_size_inches(10.5, 6.5)
plt.xlim(0,360)
ax2=ax.twiny()
labels=['21/03', '11/05', '02/07', '23/08' ,'13/10','02/12','20/01','10/03', ''] 
coordenadas=[-100, -20, 65, 140, 227, 313, 400, 486, 500]
ax2.set_xticks(coordenadas)#Calculadas las coordenadas con: https://kikimoreau.shinyapps.io/SollongCalc/ AÑO 1987
ax2.set_xticklabels(labels, rotation=70)
ax2.set_xlabel('Y[1987], dd/mm format')
ax2.set_xlabel
ax.plot(proportion_values, prior_kde(proportion_values), label='200 years')
ax.plot(proportion_values, likelihood_kde(proportion_values), label='35 years')
ax.plot(proportion_values, posterior_densities, "g--", label='Combination of both')
ax.set_xlabel('Solar longitude(º)')
ax.set_ylabel('Density')
ax.legend(bbox_to_anchor = (1,1), loc = "upper left")
plt.show()
```
