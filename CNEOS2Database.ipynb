```python
#BASE DE DATOS CNEOS(NASA), 1988-2023, 35 years period

with open("cneos.fireball.data.txt") as file_object:
    cneodata= file_object.read()
    cneo=cneodata.split(sep='\n')
    print(cneo)
```

```python
#El formato de fechas para poder utilizar la función de conversión
#debe ser 'Y/m/d', así que se reemplazará '-' por '/' en la siguiente función
```

```python
def changes(string_list):
    reem=[]
    for string in string_list :
        reem_string=string.replace('-','/')
        reem.append(reem_string)
    return reem
    
```

```python
cneook=list(map(changes,cneo))
print(cneo)
```

```python
#Unimos las sublistas
cneooklist = [''.join(sublist) for sublist in cneook]
print(cneooklist)
```

```python
#Quitamos el último valor de la lista
cneooktr=cneooklist[0:952]
cneooktr
```

```python
from datetime import datetime

#vector vacío para almacenar las fechas sin horas
onlydates= []

for datehour in cneooktr:
    #Convertir la cadena de fecha y hora en un objeto datetime
    dateobj = datetime.strptime(datehour, '%Y/%m/%d %H:%M:%S')
    #Obtener solo la fecha como cadena
    onlydate = dateobj.strftime('%Y/%m/%d')
    #Agregar la fecha sin la hora a la lista
    onlydates.append(onlydate)

print(onlydates)
```

```python
import ephem as ep #Ephemerides
import numpy as np 
import matplotlib.mlab as mlab 
import matplotlib.pyplot as plt 
import argparse
from datetime import datetime

def datetoSolarL(date): # fechas --> 'long solar' medida en la ecliptica
    s =ep.Sun()         
    s.compute(date)     #Ejecutamos la función s para calcular las efemerides en una fecha concreta
    ecl = ep.Ecliptic(s) #Transformamos las coordenadas en coordenadas eclipticas (ascensión_recta-declinación)
    ecl.lon              #Nos da la longitud ecliptica
    #print ecl.lon
    solarL=(ecl.lon/np.pi)*180 #Pasamos de radianes a grados.
    return solarL
```

```python
datacneoSolarL=list(map(datetoSolarL, cneooktr))
```

```python
rancneo=max(datacneoSolarL)-min(datacneoSolarL)
rancneo
```
